```python
# - **Jennifer Lawrence** (born 1990): Won Best Actress for "Silver Linings Playbook" (2012).
# 1. **Mahershala Ali** - Academy Awards (Oscars) for Best Supporting Actor (twice)
# 41. **Viola Davis**
# 1. **Usain Bolt** - Jamaican sprinter
# 1. Genevieve Nnaji (Nigeria - Nollywood Actress)
# - Genevieve Nnaji (Actress)
# 1. Mercy Johnson (Actress)


def get_name(txt):
    patterns = [
        "**{name}**",
        "{}. {name} ({}",
        "- {name} ({}"
    ]
    for pat in patterns:
        hits = [v['name'] for v in findall(pat, txt)]
        if len(hits) != 1:
            continue
        return hits[0]
    raise RuntimeError(f"Unable to parse name from: {txt}")

    

pattern_x = "**{name}**"
pattern_x2 = "{}. {name} ({}"
pattern_x3 = "- {name} ({}"
    
#pattern0 = "{}**{name}**{}"
test_txt_0 = '- **Jennifer Lawrence** (born 1990): Won Best Actress for "Silver Linings Playbook" (2012).'
#assert parse(pattern0, test_txt_0)['name'] == "Jennifer Lawrence"
#assert findall(pattern0, test_txt_0)['name'] == "Jennifer Lawrence"
name = [v['name'] for v in findall(pattern_x, test_txt_0)][0]
assert name == 'Jennifer Lawrence'
assert get_name(test_txt_0) == 'Jennifer Lawrence'

test_txt_1 = '1. **Mahershala Ali** - Academy Awards (Oscars) for Best Supporting Actor (twice)'
#assert parse(pattern0, test_txt_1)['name'] == "Mahershala Ali"
name = [v['name'] for v in findall(pattern_x, test_txt_1)][0]
assert name == 'Mahershala Ali'
assert get_name(test_txt_1) == 'Mahershala Ali'


#pattern1 = "{}**{name}**{}"
test_txt_2 = '# 41. **Viola Davis**'
#assert parse(pattern0, test_txt_2)['name'] == "Viola Davis"
name = [v['name'] for v in findall(pattern_x, test_txt_2)][0]
assert name == 'Viola Davis'
assert get_name(test_txt_2) == 'Viola Davis'


test_txt_3 = '1. **Usain Bolt** - Jamaican sprinter'
assert get_name(test_txt_3) == 'Usain Bolt'


test_txt_4 = '1. Genevieve Nnaji (Nigeria - Nollywood Actress)'
name = [v['name'] for v in findall(pattern_x2, test_txt_4)][0]
#print(name)
assert name == 'Genevieve Nnaji'
assert get_name(test_txt_4) == 'Genevieve Nnaji'


test_txt_5 = '- Genevieve Nnaji (Actress)'
name = [v['name'] for v in findall(pattern_x3, test_txt_5)][0]
#print(name)
assert name == 'Genevieve Nnaji'
assert get_name(test_txt_5) == 'Genevieve Nnaji'

test_txt_6 = '1. Mercy Johnson (Actress)'
name = [v['name'] for v in findall(pattern_x2, test_txt_6)][0]
#print(name)
assert name == 'Mercy Johnson'
assert get_name(test_txt_6) == 'Mercy Johnson'
```
